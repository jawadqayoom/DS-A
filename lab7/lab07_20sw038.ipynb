```python
#code for the pie chart clarity
sorted_counts = diamonds['clarity'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 90,
        counterclock = False);
plt.axis('square');
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.scatter(data=diamonds, x='carat',y='price')
plt.xlabel('carat')
plt.ylabel('price')
```

```python
sns.regplot(data=diamonds, x='carat',y='price',x_jitter=0.3,
           scatter_kws={'alpha': 1/20});
```

```python
plt.hist2d(data=diamonds, x='carat',y='price',cmin=0.5,cmap='viridis_r')
plt.colorbar()
plt.xlabel('carat')
plt.ylabel('price')
```

```python
#boxplot graph
base_color = sns.color_palette()[0]
sns.boxplot(data=diamonds, x='cut', y='price',color=base_color)
plt.xticks(rotation=15)
```

```python
sns.pointplot(data = diamonds, x = 'carat', y = 'price')
plt.ylabel('price')
plt.xticks(rotation=5)
```
